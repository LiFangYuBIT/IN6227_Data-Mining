```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# import data
train_data = 'Census Income Data Set/adult.data'
test_data = 'Census Income Data Set/adult.test'

# Add attribute names to each column
columns = ['Age','Workclass','fnlgwt','Education','Education Num','Marital Status',
           'Occupation','Relationship','Race','Sex','Capital Gain','Capital Loss',
           'Hours Per Week','Native Country','Income']
```

## 预处理

```python
# 检查，展示
train = pd.read_csv(train_data, names = columns)
train.head()
```

```python
# The first line of data is useless, skip
test = pd.read_csv(test_data, names = columns, skiprows = 1) 
test.head()     # 检查，展示
```

```python
# 原始数据集检查
print(train.info(),"\n")
print(test.info())
```

### missing data

```python
# defining function for estimating missing values in each columns
def missing_value(df):
    miss=[]
    column=df.columns
    # 假设 " ?" 是你要检测的特定值
    target_value = " ?"

    for i in column:
        # 检测整个数据框中是否包含目标值
        contains_target_value = (df[i] == target_value).any().any()
        missing = contains_target_value.sum()
        miss.append(missing)
        missing_list=pd.DataFrame(list(zip(column,miss)))
    return missing_list
```

```python
# 检查两个数据集的缺失值
print("Missing Train")
print(missing_value(train),"\n")

print("Missing Test")
print(missing_value(test))
```

```python
# 将空白数据和“ ？”进行填充
def add_miss(df):
    df.replace(" ?", pd.NaT, inplace = True)
    df.replace(" >50K", 1, inplace = True)
    df.replace(" <=50K", 0, inplace = True)
    # The below code is used soecially for test data
    df.replace(" >50K.", 1, inplace = True)
    df.replace(" <=50K.", 0, inplace = True)
    
    # 众数填充
    mode = {columns[1] : df[columns[1]].mode()[0], columns[6] : df[columns[6]].mode()[0], columns[-2] : df[columns[-2]].mode()[0]}
    df.fillna(mode, inplace = True)
    
    return df
```

```python
# train set更改空白数据并展示
train = add_miss(train)

train.head()
```

```python
# test set更改空白数据并展示
test = add_miss(test)

test.head()
```

```python
# 重新检查数据集
print(train.info(), "\n")

print(test.info())
```

```python
# 再次检查两个数据集的缺失值
print("Missing Train")
print(missing_value(train),"\n")

print("Missing Test")
print(missing_value(test))
```
