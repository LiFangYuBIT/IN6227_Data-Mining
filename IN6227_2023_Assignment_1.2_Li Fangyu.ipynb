```python
# OneHotEncode the discrete data
One_Hot_Encoder = OneHotEncoder(sparse=False)
onehot_cols = ['Workclass', 'Education', 'Marital Status', 'Occupation', 'Relationship', 'Race', 'Sex', 'Age Category', 'Hours Per Week Category']
onehot_encoded = One_Hot_Encoder.fit_transform(train[onehot_cols])
train = pd.concat([train.drop(onehot_cols, axis=1), pd.DataFrame(onehot_encoded, columns=One_Hot_Encoder.get_feature_names_out(onehot_cols))], axis=1)

onehot_encoded = One_Hot_Encoder.fit_transform(test[onehot_cols])
test = pd.concat([test.drop(onehot_cols, axis=1), pd.DataFrame(onehot_encoded, columns=One_Hot_Encoder.get_feature_names_out(onehot_cols))], axis=1)
```

## Model

```python
train_without_income = train.drop([columns[-1]], axis=1)
test_without_income = test.drop([columns[-1]], axis=1)

x_train, y_train = train_without_income.values, train[columns[-1]].values
x_test, y_test = test_without_income.values, test[columns[-1]].values

# Consider doing a mixed sampling however it dramatically decreases the accuracy
# ros = SMOTETomek(random_state=42)
# X_resampled, y_resampled = ros.fit_resample(x_train, y_train)
```

```python
def max_accuracy(df):
    if not df:
        return None  # 如果列表为空，返回 None 或适当的默认值
    return max(df)
```

### Decision Tree

```python
depths = []
for i in range(1,50):
    classifier = DecisionTreeClassifier(criterion="entropy",max_depth=i,random_state=42).fit(x_train, y_train)
    #classifier.predict(x_test)
    #score = classifier.score(x_test,y_test)
    #这里我尝试了两种方式，输出结果没有差异。后续选择使用字符更少的一种
    y_pred = classifier.predict(x_test)
    score = accuracy_score(y_test, y_pred)
    
    depths.append(score)

print(f"Highest Accuracy: {max_accuracy(depths)}") #返回最大值
print(f"Max Depth: {depths.index(max(depths))+1}") # 返回最大值的索引,因为起始是0，所以+1
    
plt.plot(range(1,50), depths)
plt.xlabel('Depth')
plt.ylabel('Accuracy')
plt.show()
```

```python
#扩大深度
depths = []
for i in range(1,100):
    classifier = DecisionTreeClassifier(criterion="entropy",max_depth=i,random_state=42).fit(x_train, y_train)
    classifier.predict(x_test)
    score = classifier.score(x_test,y_test)
    depths.append(score)

    
print(f"Highest Accuracy: {max_accuracy(depths)}") #返回最大值
print(f"Max Depth: {depths.index(max(depths))+1}") # 返回最大值的索引

plt.plot(range(1,100), depths)
plt.xlabel('Max Depth Value')
plt.ylabel('Accuracy')
plt.show()
```

#### The Best Decision Tree Model

```python
#两次索引值返回都是11
classifier = DecisionTreeClassifier(criterion="entropy",max_depth=11,random_state=42, splitter="best").fit(x_train, y_train)
y_pred = classifier.predict(x_test)

print("Accuracy of DecisionTreeClassifier:")
print(classification_report(y_test, y_pred))
print(f"Highest Accuracy: {max_accuracy(depths)}")
```

#### DT Confusion Matrix & ROC(AUC) 

```python
# 进行赋值但不影响原值
y_pred = y_pred
y_true = y_test

cm = confusion_matrix(y_true, y_pred)

plt.matshow(cm, cmap=plt.cm.Reds)

for i in range(len(cm)):
    for j in range(len(cm)):
        plt.annotate(cm[j, i], xy=(i, j), horizontalalignment='center', verticalalignment='center')

plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
auc_score = roc_auc_score(y_test, y_pred)

plt.plot(fpr, tpr, label=f'AUC = {auc_score:.2f}')
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('AUC Curve')
plt.legend()

plt.show()
```

### KNN

```python
neighbors = []
for i in range(1,60):
    knn = KNeighborsClassifier(n_neighbors=i).fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    score = accuracy_score(y_test, y_pred)
    neighbors.append(score)
    
print(f"Highest Accuracy: {max_accuracy(neighbors)}") #返回最大值
print(f"Neighbor: {neighbors.index(max(neighbors))+1}") # 返回最大值的索引

plt.plot(range(1,60), neighbors)
plt.xlabel('Range')
plt.ylabel('Accuracy')
plt.show()
```

```python
neighbors = []

for i in range(1,6):
    knn = KNeighborsClassifier(n_neighbors=28, p=i).fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    score = accuracy_score(y_test, y_pred)
    neighbors.append(score)

print(f"Highest Accuracy: {max_accuracy(neighbors)}") #返回最大值
print(f"Euclidean Distance Detric: {neighbors.index(max(neighbors))+1}") # 返回最大值的索引

plt.plot(range(1,6), neighbors)
plt.xlabel('Euclidean Distance Detric')
plt.ylabel('Accuracy')
plt.show()
```
